```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# x**2  + y**2 = r**2
r = 6
x = np.linspace(-r,r,1000)
y = np.sqrt(-x**2+r**2)
plt.plot(x, y,'b')
plt.plot(x,-y,'b')
plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=1.     #x-position of the center
v=0.5    #y-position of the center
a=2.     #radius on the x-axis
b=1.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
plt.show()
```

```python
x = np.linspace(-10, 10, 80000)
y1 = np.sqrt((6 - x**2)/3)
y2 = -(np.sqrt((6 - x**2)/3))
plt.ylim(-1.5, 1.5)
plt.xlim(-4, 4)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.linspace(-10, 10, 20000)
y1 = np.sqrt((1 - (x**2)/4**2)/3**2)
y2 = -(np.sqrt((1 - (x**2)/4**2)/3**2))
plt.ylim(-1,5, 1,5)
plt.xlim(-5, 5)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.linspace(-10, 10, 2000)
y1 = np.sqrt((1 + (x**2)/4**2)/3**2)
y2 = -(np.sqrt((1 + (x**2)/4**2)/3**2))
plt.ylim(-0,4, 0,9)
plt.xlim(-4,8, -4,4)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid (X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, color='y')
ax.scatter(0, 0, 0, 'z', 50)
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid (X, Y)
Z = 2*X + 5*Y
ax.plot_surface(X, Y, Z, color='y')
ax.plot_surface(X, Y + 5, Z, color='m')
ax.scatter(0, 0, 0, 'z', 50)
show()
```
